# Preparing Dataset

```python
data = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
data.head(5)
```

```python
data['Result'] = diabetes.target
```

```python
# Dataset separation in independent and depedent

X = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
X.shape,y.shape
```

```python
X.head()
```

```python
y.head()
```

```python
#Train test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
# standardize the data to have our global_minima wjile gradient descent while normalizing the data

from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X_train = scalar.fit_transform(X_train)
```

```python
X_test = scalar.transform(X_test)
```

```python
import pickle

pickle.dump(scalar, open('scalar.pkl','wb'))
```

```python
X_train
```

```python
X_test
```

# Train the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
```

```python
print(reg.intercept_)
```

```python
reg.get_params()
```

```python
reg_pred = reg.predict(X_test)
```

```python
reg_pred
```

```python
plt.scatter(reg_pred,y_test)
```

```python
#Residuals

residuals = y_test - reg_pred
```

```python
residuals
```

```python
sns.displot(residuals, kind='kde')
```

```python
#scatter plot with residuals and prediction

plt.scatter(residuals, reg_pred)
```

```python
# parameters to check model

from sklearn.metrics import mean_absolute_error, mean_squared_error

print(mean_squared_error(y_test, reg_pred))
print(mean_absolute_error(y_test, reg_pred))
```

```python
print(np.sqrt(mean_squared_error(y_test, reg_pred)))
```

```python
from sklearn.metrics import r2_score

print(r2_score(y_test, reg_pred))
```
